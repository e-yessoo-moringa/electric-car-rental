# Data Cleaning

## Step 1: Finding and counting missing data

```python
# Counting data in the columns

carrental.count()
```

```python
# Counting all null data

np.count_nonzero(carrental.isnull())
```

### a) Recoding missing values

We will not drop missing values because ----
instead we will recoded missing values with NA because ---




```python
# Replacing all missing values with NA

carrental_recoded = carrental.fillna("NA")
carrental_recoded
```

```python
# Checking for missing values in recoded dataset

np.count_nonzero(carrental_recoded.isnull())
```

```python
# Verifying that the shape of the dataset is unchanged

carrental_recoded.shape
```

## Step 2 : Data Validity

```python
# Dropping irrelevant data

carrental_recoded.drop(["Cars" , "Charge Slots" , "Charging Status" , "Displayed comment" , "Kind", "Geo point" , "Public name" , "Scheduled at" , "Slots", "Station type" , "Status" , "Subscription status"], axis = 1, inplace = True) 
carrental_recoded
```

```python
# Verifying that only 12 columns were removed

carrental_recoded.shape
```

## Step 3 : Accuracy

```python
# Checking for duplicates

carrental_recoded.duplicated().sum()
```

```python
carrental_recoded.duplicated()
```

## Step 4 : Uniformity

```python
carrental_recoded.columns = carrental_recoded.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(carrental_recoded)
```

```python
carrental_recoded
```

```python
carrental_recoded.info()
```

## Step 5 : Data validation

Keeping data for stations that are operational only.

```python
# Only keeping data for stations that are operational

carrental_cleaned = carrental_recoded[carrental_recoded.rental_status =='operational']
carrental_cleaned
```

```python
carrental_cleaned.shape
```

### Question 1
Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
# Filtering data for the city of Paris only, then we filtered for data when the “bluecar_counter” was zero. 

# We assumed that when “bluecar_counter” is zero, it implies that there were no cars in the station. 
# So, these were the times when the maximum number of cars were rented out. 


carrental_paris_allpickedup = carrental_cleaned[(carrental_cleaned['bluecar_counter']  == 0) & (carrental_cleaned['city'] == "Paris")]
carrental_paris_allpickedup
```

```python
carrental_paris_allpickedup['hour'].mean()
```

```python
carrental_paris_allpickedup['hour'].mode()
```

```python
carrental_paris_allpickedup['bluecar_counter'].groupby(carrental_paris_allpickedup['hour']).count().sort_values(ascending=False)
```
